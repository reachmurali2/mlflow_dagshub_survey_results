```python
# Range of customer age 
age_range = df.age.max()-df.age.min()

# 2. Mean,Median, mode of amount spent
amount_mean = df['amount_spend'].mean()
amount_median =df['amount_spend'].median()
amount_mode = df['amount_spend'].mode()

# 3.Variance and standard Deviation of time spent on site.
time_variance = round(df['time_spent_on_site'].var(),2)
time_std = round(df['time_spent_on_site'].std(),2)

# 4. Interquartile Range (IQR) of amount spent 
q3 = df['amount_spend'].quantile(0.75)
q1 = df['amount_spend'].quantile(0.25)
amount_iqr = q3-q1

# 5. correlation between no. of purchases made and time spent on site
purchase_time_corr = df['no_of_purchases'].corr(df['time_spent_on_site'])
purchase_time_corr = round(purchase_time_corr,2)

# 6.z_score for customer who spends $450
z_score_450 = (450-amount_mean)/df['amount_spend'].std()
z_score_450 = round(z_score_450,2)

# 7. Skewness of amount spend
skewness_of_amount_spent = df['amount_spend'].skew()

# 8. Probability of >5 purchases or spending >$300
prob_over_5_purchases = (df['no_of_purchases']>5).mean()
prob_over_300 = (df['amount_spend']>300).mean()
prob_both = ((df['no_of_purchases']>5) & (df['amount_spend']>300)).mean()
prob_either = prob_over_5_purchases + prob_over_300 - prob_both

# 9. 95% Confidence Interval for amount spent
import scipy.stats as stats
z_critical = stats.norm.ppf(q=0.975)   # 95% CI
margin_of_error = z_critical*(df['amount_spend'].std()/len(df)**0.5)
ci_lower = amount_mean-margin_of_error
ci_upper = amount_mean+margin_of_error
ci_lower = round(ci_lower,0)
ci_upper = round(ci_upper,0)

# 10.Probability of high conversion rate(>80%) among customers with 5+ purchases

high_purch_conv = df[df['no_of_purchases']>=5]
prob_high_conversion=(high_purch_conv['cross_sell_conversion_rate']>.8).mean()
prob_high_conversion = round(prob_high_conversion,3)

{
    "1.Age Range":age_range,
    "2.Mean Amount Spent": amount_mean,
    "2.Median Amount Spent": amount_median,
    "2.Mode Amount Spent": amount_mode,
    "3.Variance time on site":time_variance,
    "3.Std deviation of time on site":time_std,
    "4.IQR for amount spent": amount_iqr,
    "5.Correlation (Purchase vs Time)":purchase_time_corr,
    "6.z_score for $450_spend":z_score_450,
    "7.Skewness of Amount Spent":skewness_of_amount_spent,
    "8.P(X>5 Purchases or spent >$300)":prob_either,
    "9.95% CI for Amount Spent":(ci_lower,ci_upper),
    "10.: P(High Conversion|>=5 Purchases)":prob_high_conversion
    

}
```

```python
"""
10.What is the probability that a customer who makes 5 or more purchases in a
month will have a high cross-sell conversion rate (greater than 80%)?

Hint:  
step 1 : Find the proportion of customers with high cross-sell conversion rate (greater than 80%)
step 2 : Find the proportion of customers who make 5 or more purchases in a month
step 3 : Calculate the proportion of high cross-sell conversion who make 5 or more purchases.
setp 4 : Use Baye's theorem to calculate the probability of high cross-sell conversion given 5 or more purchases
"""
# step 1 : P(High Cross-sell Conversion) 
p_high_conversion=len(df[df['cross_sell_conversion_rate']>0.8])/len(df)

# step2 : P(>=5 Purchases)
p_5_or_more_purchases = len(df[df['no_of_purchases']>=5])/len(df)

# Step 3: P(>=5 Purchases and High Cross-Sell Conversion)

both_condition = df[(df['cross_sell_conversion_rate']>0.8) & (df['no_of_purchases']>=5)]
p_both = len(both_condition)/len(df)

# step 4: Baye's therem : P(high conversion | >=5 Purchases)
# P (A|B) = P(A ∩ B) / P(B)
p_high_conversion_given_5purchases = p_both/p_5_or_more_purchases

p_high_conversion, p_5_or_more_purchases, p_both, p_high_conversion_given_5purchases
```
